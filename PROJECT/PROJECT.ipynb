## Il grafico seguente rendendolo rettangolare e non quadrato viene come quello sulla scheda del progetto? Capisci come stiracchiare il grafico

```r
counts=read.delim("B19036_AmCsCo_20180316.dat", header = TRUE)
log.counts=log2(as.numeric(counts[,1]))
par(mfrow=c(1,1))
plot(log.counts[log.counts>0], type = 'l', col='red', main = 'Am−Cs−Co spectra', xlab = 'ADC channel', ylab = "log2(counts)",cex.main = 1.5)
```
